```python
df['Day']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
#df['Day']=df['Date_of_Journey'].apply(lambda x:x.split('/')[0]
#df['Month']=df['Date_of_Journey'].apply(lambda x:x.split('/')[1]
#df['Year']=df['Date_of_Journey'].apply(lambda x:x.split('/')[2]
```

```python
df['Arr_Hour']=df['Arrival_Time'].str.split(':').str[0].astype(int)
df['Arr_Mins']=df['Arrival_Time'].str.split(':').str[1].astype(int)
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df['Dep_Hour']=df['Dep_Time'].str.split(':').str[0].astype(int)
df['Dep_Mins']=df['Dep_Time'].str.split(':').str[1].astype(int)
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize and fit LabelEncoder
le = LabelEncoder()
df['Total_Stops'] = le.fit_transform(df['Total_Stops'])
df['Additional_Info'] = le.fit_transform(df['Additional_Info'])
df['Airline'] = le.fit_transform(df['Airline'])
df['Source'] = le.fit_transform(df['Source'])
df['Destination'] = le.fit_transform(df['Destination'])
print("Category Mapping:", le.classes_)
```
