```python
df_country=pd.read_excel('/Users/jitikagupta/Downloads/Zomatodataset/Country-Code.xlsx')
```

```python
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating_Count'})
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

First one â†’ messy list, counts hidden in the index.
Second one â†’ clean table, counts shown in a new column called Rating Count.
Second one (DataFrame with Rating Count)
Much cleaner.
Columns are explicit, so you can:
Easily filter (e.g., df[df['Rating Count'] > 100])
Plot with seaborn/matplotlib/plotly
Merge with other tables
Export to CSV/Excel

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating_Count',hue='Rating color',data=ratings,palette=['Blue','Red','Orange','Yellow','Green','Green'])
plt.show()
```

```python
#count plot
sns.countplot(x='Rating color',data=ratings,palette=['Blue','Red','Orange','Yellow','Green','Green'])
```
